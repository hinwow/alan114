```python
train_data=pd.read_csv('/kaggle/input/moai/Iris_train_data.csv')
test_data=pd.read_csv('/kaggle/input/moai/Iris_test_data.csv')
a=preprocessing.LabelEncoder()
train_data['Species']= a.fit_transform(train_data['Species'])
X_train=train_data.drop(['Species','Id'],axis=1)
y_train=train_data.loc[:,'Species']
X_test=test_data.drop(['Id'],axis=1)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # 用训练集拟合标准化器
X_test_scaled = scaler.transform(X_test)  # 测试集用相同的标准化参数
```

```python
# 初始化逻辑回归模型（L2正则化，C=1.0表示正则化强度）
model = LogisticRegression(penalty='l2', C=1.0, max_iter=1000)  # 增加max_iter确保收敛

# 训练模型
model.fit(X_train_scaled, y_train)
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)  # 随机打乱数据
scores = cross_val_score(model, X_train_scaled, y_train, cv=kfold, scoring='accuracy')
print(f"各折准确率: {scores}")
print(f"平均准确率: {scores.mean():.4f}（标准差: {scores.std():.4f}）")
```

```python
y_pred_prob = model.predict_proba(X_test_scaled)[:, 1]  # 正类（良性）的概率
y_pred = model.predict(X_test_scaled)  # 类别标签（0或1）
```
